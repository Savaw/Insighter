```python
import numpy as np
import matplotlib.pyplot as plt
```

### Setup parameters

```python
w = 2 * np.pi
z = 0.2 
x_0 = [0, 2] # Initial condition

dt = 0.01
T = 10
```

### Initialize states and transition matrix

```python
A = np.array([[0, 1], [-w**2, -2 * w * z]])

n_timestamps = int(np.ceil(T/dt) + 1)
timestamps = np.linspace(0, T, n_timestamps)

# List of states
# Each state is a vector of [x v]^T
states = np.zeros((n_timestamps, 2))
states[0] = np.array(x_0)
```

### Simulate using forward Euler method

```python
t = 0
k = 0
while t+dt < T:
    t += dt 
    k += 1
    states[k] = states[k-1] + dt * A @ states[k-1]
    
```

### Plots

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(8, 3), constrained_layout=True)
# fig.subplots_adjust(right=1.5)

axes[0].set_xlabel("Time")
axes[0].set_ylabel("Position")
axes[0].plot(timestamps, states[:, 0], color='blue')

axes[1].set_xlabel("Time")
axes[1].set_ylabel("Velocity")
axes[1].plot(timestamps, states[:, 1], color='purple')

axes[2].set_xlabel("Position")
axes[2].set_ylabel("Velocity")
axes[2].plot(states[:, 0], states[:, 1], color='orange')

plt.show()
```
